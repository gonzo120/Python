```python
import pandas as pd
from string import ascii_uppercase as alfabeto
import pickle
```

```python
todas_tablas = pd.read_html('https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')
todas_tablas
```

```python
# Grupos A - H
# 11 -> 7X8 +11 =67
#Utilizamos este llamado para visualizar la posicion de las tablas del mundial
todas_tablas[12]
todas_tablas[19]
todas_tablas[26]
```

```python
#Visualizamos la columna 1 de la tabla 12 del grupo A
#Esto para poder entender como cambiarlo a todas las tablas
todas_tablas[12].columns[10]
```

```python
#Aqui notamos como relacionamos la letra del alfabeto con la tabla de cada equipo que esta por numero
#la idea es convertir el numero a letra para visualizar mejor las tablas 
#Funcion zip()combina dos listas en pares ordenados
for letra, i in zip(alfabeto, range(12,68,7)):
    print(letra, i)
```

```python
#range() nos permite hacer una lista con ciertos elementos
#cambiamos el nombre de las columnas
    #inplace sirve para guardar los cambios
    #pop sirve para eliminar columnas
#Incluimos el alfabeto en las tablas
dict_tablas = {}
for letra, i in zip(alfabeto, range(12,68,7)):
    df = todas_tablas[i]
    df.rename(columns={df.columns[1]:'Team'},inplace=True)
    #df.pop('Qualification')
    dict_tablas[f'Grupo {letra}'] = df
    print(df)
#Tenemos todas las tablas
#Extraemos todas las tablas

```

```python
#Devuelve todos los elementos
dict_tablas.keys()
```

```python
dict_tablas['Grupo A']
```

```python
#===========================================================================================================================
#Se finaliza la extracción de datos
```

```python
#En resumen, este código guarda el contenido del diccionario dict_tablas
#en un archivo llamado 'dict_table' utilizando la biblioteca pickle. 
#La declaración with open() se utiliza para asegurar que el archivo se cierre correctamente una vez que se haya completado la operación de escritura.
with open ('dict_table','wb') as output:
    pickle.dump(dict_tablas,output)
```
