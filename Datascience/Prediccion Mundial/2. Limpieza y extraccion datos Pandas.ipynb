```python
import pandas as pd
import numpy as np
```

# Limpiando Data


```python
df_data_historica = pd.read_csv('fifa_worldcup_historical_data.csv')
df_fixture = pd.read_csv('fifa_worldcup_fixture1.csv')
df_data_historica
```

# Limpiamos los datos df_fixture

```python
df_fixture['home'] = df_fixture['home'].str.strip() #strip sirve para limpiar los espacios en blanco
df_fixture['away'] = df_fixture['away'].str.strip() #strip sirve para limpiar los espacios en blanco

df_fixture
```

# Limpiamos Df_Data_Historica

```python
df_data_historica[df_data_historica['home'].isnull()]
df_data_historica.dropna(inplace=True)
#buscamos los partidos que son de suecia
df_data_historica[df_data_historica['home'].str.contains('Sweden') & 
                  df_data_historica['away'].str.contains('Austria')]
# encontramos un partido que el score es una cadena de texto que nos afectara 
df_data_historica
```

```python
#Vamos a eliminar el dato de nuestro dara frame
index_eliminar = df_data_historica[df_data_historica['home'].str.contains('Sweden') & 
                  df_data_historica['away'].str.contains('Austria')].index
index_eliminar
```

```python
df_data_historica.drop(index=index_eliminar, inplace=True)
```

```python
df_data_historica[df_data_historica['home'].str.contains('Sweden') & 
                  df_data_historica['away'].str.contains('Austria')]
#Verificamos que el dato ya no este en nuestro dataframe
```

```python
df_data_historica[df_data_historica['score'].str.contains('[^\d-]')]
##buscamos los partidos que tengan (a.e.t) para eliminar y dejar solo el marcador
```

```python

df_data_historica['score'] = df_data_historica['score'].str.replace('\s*\([^)]*\)', '', regex=True)


df_data_historica
```

```python
df_data_historica['score'] = df_data_historica['score'].astype(str)

df_data_historica[df_data_historica['score'].str.contains('a.e.t')]

#Aqui ya tenemos los score sin las siglas a.e.t
```

```python
df_data_historica['home'] = df_data_historica['home'].str.strip() #strip sirve para limpiar los espacios en blanco
df_data_historica['away'] = df_data_historica['away'].str.strip() #strip sirve para limpiar los espacios en blanco

df_data_historica
```

# se separa los datos de score

```python
df_data_historica[['HomeGoals','AwayGoals']] = df_data_historica['score'].str.split('–', expand=True)

df_data_historica
```

```python
df_data_historica.drop('score', axis=1, inplace=True)
```

```python
df_data_historica.rename(columns={'home':'HomeTeam','away':'AwayTeam','year':'Year'},inplace=True)
df_data_historica[df_data_historica]
```

```python
##Verificamos el tipo  de dato

df_data_historica.dtypes
```

```python


# Verificar si los valores son numéricos utilizando pd.to_numeric()
numeric_values = pd.to_numeric(df_data_historica['HomeGoals'], errors='coerce')
non_numeric_values = df_data_historica['HomeGoals'][numeric_values.isna()]

# Mostrar los valores no numéricos
print("Valores no numéricos:")
print(non_numeric_values)

# Verificar si los valores son numéricos utilizando pd.Series.str.isnumeric()
non_numeric_values = df_data_historica['HomeGoals'][~df_data_historica['HomeGoals'].str.isnumeric()]

# Mostrar los valores no numéricos
print("Valores no numéricos:")
print(non_numeric_values)
fila = df_data_historica.loc[df_data_historica['HomeGoals'] == '3-3']
print(fila)
```

```python


# Reemplazar el valor '3-3' manualmente para la fila 358
df_data_historica.loc[358, 'HomeGoals'] = '3'
df_data_historica.loc[358, 'AwayGoals'] = '3'

# Convertir las columnas a tipo numérico
df_data_historica[['HomeGoals', 'AwayGoals']] = df_data_historica[['HomeGoals', 'AwayGoals']].apply(pd.to_numeric, errors='coerce')

# Verificar los valores no numéricos nuevamente
non_numeric_values = df_data_historica['HomeGoals'][df_data_historica['HomeGoals'].isna()]
print("Valores no numéricos:")
print(non_numeric_values)

# Imprimir el DataFrame actualizado
#Verificamos que ya no este el error
print(df_data_historica)
fila = df_data_historica.loc[df_data_historica['HomeGoals'] == '3-3']
print(fila)
```

```python
df_data_historica = df_data_historica.astype({'HomeGoals': int, 'AwayGoals': int, 'Year':int})
df_data_historica.dtypes
```

# Exportamos los dataframes limpios

```python
df_data_historica.to_csv('clean_fifa_worldcup_marches.csv', index = False )
df_fixture.to_csv('clean_fifa_worldcup_fixture.csv', index = False )
```
