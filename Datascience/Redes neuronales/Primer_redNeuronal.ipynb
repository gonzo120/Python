# primer red neuronal 
en este programa se crea la red neuronal para hacer el calculo de
celsius a fahrenheit

```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40,-10,0,8,15,22,38],dtype=float)
fahrenheit = np.array([-40,14,32,46,59,72,100],dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1), #Adam permite ajustar pesos y sesgos para que aprend
    loss='mean_squared_error'#
)
```

```python
print("comenzando entrenamiento....")
historial =modelo.fit(celsius,fahrenheit, epochs=1000,verbose=False)
#epochs=da cada vuelta que da para entrenar el modelo
#verbose=false para que no nos muestre los resutlados
print("Modelo entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel(" # epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("veamos que tan bueno es el modelo")
resultado = modelo.predict([100.0])
print("El resultado es" + str(resultado) + "fahrenheit!")
```

```python
print("variables internas del modelo")
print(capa.get_weights())
#primer dato pesos, segundo dato sesgo
```
