 # Creaci√≥n de varios Histogramas por medio d euna funcion
 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_atletas = pd.read_csv('C:/Users/sao/Desktop/Python/Datascience/datasets/categorias de corredores.csv', index_col=0)


#Creamos el histograma de la variable tiempo

plt.figure(0)
plt.hist(df_atletas['Tiempo'], 15, color="yellow", ec="black")
plt.title("Histograma Tiempo")

#creamos el histograma de la variable edad


plt.figure(1)
plt.hist(df_atletas['Edad'], 15, color="yellow", ec="black")
plt.title("Histograma Edad")
```

se creara una funcion para formar histogramas

```python
def crear_histograma(variable,j):
    plt.figure(j)
    plt.hist(df_atletas[variable], 15, color="yellow", ec="black")
    plt.title("Histograma"+variable)
```

```python
crear_histograma("Edad")
```

se utiliza una misma funcion para crear varios histogramas

```python
columnas=["Edad","Tiempo"]
j=0 #Inicializamos una variable a utilizar para separar los histogramas
for i in columnas:
    crear_histograma(i,j)#Mandamos a llamar las funciones
    j += 1 #incrementaremos la variable j
```

```python


def crear_histograma1(Columnas,j,color):
    plt.figure(j)
    plt.hist(df_atletas[Columnas], 15, color=color, ec="black")
    plt.title("histograma"+Columnas)

Columnas = ["Lugar", "Tiempo", "Edad"]
colores= ["yellow", "green", "blue"]
j=0
k=0

for i in Columnas:
    crear_histograma1(i,j,colores[k])
    j += 1
    k += 1
```

```python
def crear_histograma2(columnas,j,color):
    plt.figure(j)
    plt.hist(df_atletas[columnas],15,color=color,ec="black")
    plt.title("Histograma"+" "+columnas)

columnas=["Lugar","Edad","Tiempo"]
color=["yellow","green","black"]

j=0
k=0

for i in columnas:
    crear_histograma2(i,j,color[k])
    j += 1
    k += 1
```
